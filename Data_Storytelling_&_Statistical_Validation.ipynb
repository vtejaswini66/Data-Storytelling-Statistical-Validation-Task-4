```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
np.random.seed(42)

old_design = np.random.normal(loc=0.12, scale=0.02, size=100)
new_design = np.random.normal(loc=0.15, scale=0.02, size=100)

df = pd.DataFrame({
    "Old_Design": old_design,
    "New_Design": new_design
})

df.head()
```

```python
plt.figure(figsize=(6,4))

sns.kdeplot(old_design, label="Old Design")
sns.kdeplot(new_design, label="New Design")

plt.title("Conversion Rate Distribution")
plt.legend()

plt.show()
```

H0 (Null Hypothesis):
New design does NOT improve conversion rate

H1 (Alternative Hypothesis):
New design improves conversion rate

```python
t_stat, p_value = stats.ttest_ind(new_design, old_design)

print("T-statistic:", t_stat)
print("P-value:", p_value)
```

```python
alpha = 0.05

if p_value < alpha:
    print("Reject Null Hypothesis")
    print("New website design significantly improves conversions")
else:
    print("Fail to reject Null Hypothesis")
```

```python
print("Task 4 Completed")

print("Old Conversion Mean:", old_design.mean())
print("New Conversion Mean:", new_design.mean())

print("P-value:", p_value)
```
